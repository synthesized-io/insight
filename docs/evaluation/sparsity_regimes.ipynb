<h1><center>5. Uniform Distribution</center></h1>

```python
def plot_two_dists(data1: pd.DataFrame, data2: pd.DataFrame, label1='Original', label2='Missing', title=None):
    plt.figure(figsize=(10, 6))
    sns.distplot(data1['x'], color=sns.color_palette()[0], label=label1)
    sns.distplot(data2['x'].dropna(), color=sns.color_palette()[1], label=label2)
    if title:
        plt.title(title)
    plt.legend()
    plt.show()
    
```

```python
data = syn_dist.create_uniform_categorical(n_classes=100, size=10000)
plot_missing_patterns(data, plot_function=plot_two_dists, mechanisms=['MCAR','MNAR'])
```

```python
results = dict()
for mechanism in ['MCAR', 'MNAR']:
    results[mechanism] = synthesize_and_plot_results(
        data, mechanism, n_iter=2500, std_noise = 2., start=25, end=100, step=5, n_experiments=3
    )
```

```python
plt.figure(figsize=[12,8])
plt.plot([x/100 for x in range(25, 101, 5)], results['MCAR'][0], label='MCAR')
plt.plot([x/100 for x in range(25, 101, 5)], results['MNAR'][0], label='MNAR')
plt.legend()
plt.title('KS Distance')
plt.xlabel('Non-Missing Ratio')
plt.ylabel('Distance')
plt.show()
```

#Mixed Continuous and Categorical

```python
def plot_two_dists(data1: pd.DataFrame, data2: pd.DataFrame, label1='Original', label2='Missing', title=None):
    df_plot = pd.concat((data1.assign(origin=label1) , data2.assign(origin=label2)))
    plt.figure(figsize=(10, 6))
    sns.violinplot(x="x", y="y", hue='origin', split=True, data=df_plot, title=title)
    if title:
        plt.title(title)
    plt.legend()
    plt.show()
    
```

```python
data = syn_dist.create_mixed_continuous_categorical(n_classes=10, size=10000, prior_sd=2.)
plot_missing_patterns(data, plot_function=plot_two_dists)
```

```python
results = dict()
for mechanism in ['MCAR', 'MAR', 'MNAR']:
    results[mechanism] = synthesize_and_plot_results(
        data, mechanism, n_iter=2500, std_noise = 2., start=25, end=100, step=5, n_experiments=3
    )
```
